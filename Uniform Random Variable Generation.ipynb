```python
# Distribution of Uniformly drawn random variables
rand_X = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
rand_Y = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
RV = np.sqrt(-2*np.log(rand_X))*np.sin(2*np.pi*rand_Y)

# Plot distribution of Random Variable
plt.hist(RV)
plt.show()
```

```python
# Distribution of Uniformly drawn random variables
rand_X = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
rand_Y = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
RV = rand_X / (rand_X - rand_Y)

# Plot distribution of Random Variable
plt.hist(RV, bins=50)
plt.show()
```

```python
# Distribution of Uniformly drawn random variables
lamb = 1
beta = 5
rand_X = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
RV = (1 / lamb) * (-np.log(1 - rand_X))**(1/beta)

# Plot distribution of Random Variable
plt.hist(RV, bins=50)
plt.show()
```

Use Monte Carlo simulation to estimate the expected value of the loan that will be needed to enable Joey to go to college next year.

```python
# Expected value of a loan given a persons college costs and income streams

# Costs
tuition = 8400
dormitory = 5400
meals = np.array([np.random.uniform(900, 1350) for i in range(0, 10000)])
entertainment = np.array([np.random.uniform(600, 1200) for i in range(0, 10000)])
transportation = np.array([np.random.uniform(200, 600) for i in range(0, 10000)])
books = np.array([np.random.uniform(400, 800) for i in range(0, 10000)])

# Income
scholarship = 3000
parents = 4000
waiting = np.array([np.random.uniform(3000, 5000) for i in range(0, 10000)])
library = np.array([np.random.uniform(2000, 3000) for i in range(0, 10000)])

# Simulation
total_cost = tuition + dormitory + meals + entertainment + transportation + books
total_income = scholarship + parents + waiting + library
loan_needed = np.mean(total_cost - total_income)

print("Average total cost:", np.mean(total_cost))
print("Average income:", np.mean(total_income))
print("Loan size needed:", loan_needed)
```
