```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Distribution of Uniformly drawn random variables
rand_unif_nums = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
RV = -np.log(rand_unif_nums)

# Plot distribution of Random Variable
plt.hist(RV)
plt.show()
```

Distribution is similar to exponential distribution

```python
# Distribution of Uniformly drawn random variables
rand_X = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
rand_Y = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
RV = np.sqrt(-2*np.log(rand_X))*np.sin(2*np.pi*rand_Y)

# Plot distribution of Random Variable
plt.hist(RV)
plt.show()
```

Distribution is similar to a normal distribution

```python
# Distribution of Uniformly drawn random variables
rand_X = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
rand_Y = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
RV = rand_X / (rand_X - rand_Y)

# Plot distribution of Random Variable
plt.hist(RV, bins=50)
plt.show()
```

```python
# Distribution of Uniformly drawn random variables
lamb = 1
beta = 5
rand_X = np.array([np.random.uniform(0, 1) for i in range(0, 10000)])
RV = (1 / lamb) * (-np.log(1 - rand_X))**(1/beta)

# Plot distribution of Random Variable
plt.hist(RV, bins=50)
plt.show()
```

Distribution is similar to a Weibull distribution
